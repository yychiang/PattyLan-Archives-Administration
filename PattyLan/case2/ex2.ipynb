```python
file1 = pd.read_csv('ex2.csv')
print(file1)
```

```python
reg=file1.loc[ 0 , '單位（12級）' ]
print(reg)
counter=0
```

```python
for i in range(len(file1)):
    regnew =  regnew = file1.loc[ i , '單位（12級）' ]
    if (regnew == reg):
        counter=counter+1
        file1.at[i,'單位序'] = counter 
    else:
        reg = regnew
        counter = 1
        if not(pd.isna(reg)):
            file1.at[i,'單位序'] = counter
```
